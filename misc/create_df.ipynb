```python
import os 
import cv2
import pandas as pd
import re

data_path = '/arc/project/st-anaray02-1/skumar40/petraw_data/Training/'
```

```python
#### FOR FRAMES AND SEGMENTATION ####

data_typ = ['Frames']
numbers = re.compile(r'(\d+)')

def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts

for typ in data_typ:
    file_path = []
    folder_name = []
    main_folder_path = data_path + typ
    for f in sorted(os.listdir(main_folder_path)):
        sub_folder_list_path = [main_folder_path+'/'+f+'/'+i for i in sorted(os.listdir(main_folder_path + '/' + f), key=numericalSort)]
        folder_list = [f] * len(sub_folder_list_path)

        file_path.extend(sub_folder_list_path)
        folder_name.extend(folder_list)

    df = pd.DataFrame({'file_path': file_path, 'folder_name': folder_name})
    df.to_csv('./dataset_path/' + typ + '.csv')
```
